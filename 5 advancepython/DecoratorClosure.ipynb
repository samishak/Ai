```python
def outer_function(x):
    def inner_function(y):
        return x + y   # x is remembered
    return inner_function
```

```python
add_five = outer_function(5)

print(add_five(3))
print(add_five(10))
```

```python
def my_decorator(func):
    def wrapper():
        print("Before function runs")
        func()
        print("After function runs")
    return wrapper
```

```python
@my_decorator
def say_hello():
    print("Hello!")
```

```python
say_hello()
```
