```python
#accessing data from dataframe
df['Name']
type(df['Name']) #Series
```

```python
df.loc[0]  #first row as Series
type(df.loc[0]) #Series
```

```python
df.iloc[0]  #first row as Series
type(df.iloc[0]) #Series
df.loc[0,'Name']  #specific value
type(df.loc[0,'Name']) #string
df.iloc[0,0]  #specific value
type(df.iloc[0,0]) #string
# iloc is integer-location based indexing for selection by position whereas loc is label-based indexing for selection by label.
#loc includes the last element of the range specified while iloc excludes it.
df[['Name','City']]  #multiple columns
type(df[['Name','City']]) #DataFrame
```

```python
df.at[1,'Age']  #specific value
type(df.at[1,'Age']) #int
df.iat[1,1]  #specific value
type(df.iat[1,1]) #int
```

```python
#data manipulation with dataframe
df['Salary']=[70000,80000,90000]
df
```

```python
#to remove a column
df.drop('Salary',axis=1,inplace=True)
df
```

```python
#add age to a column
df['Age']=df['Age']+5
df
```

```python
df.drop(0,axis=0,inplace=True)
df 
```

```python
#display the data types of each column 
df.dtypes
df.info()
#display summary statistics of numerical columns
df.describe()
df.describe(include='all')
#sort the dataframe by age in descending order
df.sort_values(by='Age',ascending=False)
df.sort_values(by='Name')
df.sort_values(by=['Age','Name'],ascending=[False,True])
#filter rows where age is greater than 30
df[df['Age']>30]
df.loc[df['Age']>30]
df[df['City']=='Chicago']
df.loc[df['City']=='Chicago']
#filter rows where age is greater than 30 and city is Chicago
df[(df['Age']>30) & (df['City']=='Chicago')]
df.loc[(df['Age']>30) & (df['City']=='Chicago')]
#filter rows where age is greater than 30 or city is Chicago
df[(df['Age']>30) | (df['City']=='Chicago')]
df.loc[(df['Age']>30) | (df['City']=='Chicago')]
#reset the index of the dataframe
df.reset_index(drop=True,inplace=True)
df
```
