```python
# filling missing values with the column mode (most frequent city)
df['city_fillNA'] = df['City'].fillna(df['City'].mode()[0])
df.drop('Sales_fillNA', axis=1, inplace=True)
df
```

```python
df.dtypes
```

```python
##renaming columns
df.rename(columns={'City':'Citynew', 'Country':'Countrynew'}, inplace=True)
df.head()
```

```python
#change column data type
df['indexNew'] = df['Index'].astype(float)
df.head()
df.dtypes
#df.head() is for displaying the first 5 rows of the dataframe
```

```python
#create a column with random values
import numpy as np
df['RandomValues'] = np.random.randint(1,100,size=len(df))
df.head()
```

```python
#change column data type
df['NewRanvalues'] = df['RandomValues'].astype(float)
df.head()
```

```python
#data aggregating and grouping
df.head()
```

```python
#data aggregating and grouping
df_grouped = df.groupby('Countrynew').agg({'NewRanvalues':'sum', 'Index':'mean'})
df_grouped.reset_index(inplace=True)
df_grouped
```

```python
# use the correct column name 'Customer Id' (capital I)
grouped_sum = df.groupby(['Customer Id', 'Countrynew'])['NewRanvalues'].sum()
print(grouped_sum)
```

```python
##aggregating multiple functions
grouped_agg=df.groupby('Countrynew').agg({'NewRanvalues':['sum','mean','max'], 'Index':['mean','min']})
grouped_agg
```

```python
#marging and joining dataframes
df1=pd.DataFrame({'Key':['A','B','C'],'Value1':[10,20,30]})
df2=pd.DataFrame({'Key':['A','B','D'],'Value2':[40,50,60]})
```

```python
df1
```

```python
## merge dataframes on 'Key' column
merged_df = pd.merge(df1, df2, on='Key', how='outer')
merged_df
#difference between inner and outer join is inner join returns only the common keys in both dataframes 
# while outer join returns all keys from both dataframes filling missing values with NaN
#and left join returns all keys from the left dataframe and matching keys from the right dataframe
#right join returns all keys from the right dataframe and matching keys from the left dataframe
```
