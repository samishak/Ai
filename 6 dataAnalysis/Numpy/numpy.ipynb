```python
import numpy as np
#create array using numpy
# Create a 1D array
array_1d = np.array([1, 2, 3, 4, 5])
print("1D Array:", array_1d)
print("type of array:", type(array_1d))
print("Shape:", array_1d.shape)
```

```python
arr=np.array([[1,2,3],[4,5,6]])
print(arr.sum()) #sum of all elements
print("shape:",arr.shape)
print("number of dimensions:",arr.ndim)
print("data type:",arr.dtype)
print("size:",arr.size) #total number of elements
print("type:",type(arr))
print("sum of columns:",arr.sum(axis=0)) #sum of each column
print("sum of rows:",arr.sum(axis=1)) #sum of each row
```

```python
#numpy vectorized operations
a=np.array([1,2,3,4])
b=np.array([5,6,7,8])

print("Addition:",a+b)
print("Subtraction:",b-a)
print("Multiplication:",a*b)
print("Division:",b/a)
print("Exponentiation:",a**2)
print("Square Root:",np.sqrt(a))
print("Sine:",np.sin(a))
print("Cosine:",np.cos(a))
print("Logarithm:",np.log(a))
print("Exponential:",np.exp(a))
print("Maximum:",np.maximum(a,b))
print("Minimum:",np.minimum(a,b))
print("Mean of a:",np.mean(a))
print("Standard Deviation of a:",np.std(a))
print("Dot Product:",np.dot(a,b))
# --- IGNORE ---
```

```python
#statistics with numpy
data = np.array([12, 15, 14, 10, 18, 20, 22, 19, 17, 16])
mean = np.mean(data)
median = np.median(data)
std_dev = np.std(data)
variance = np.var(data)
print("Mean:", mean)
print("Median:", median)
print("Standard Deviation:", std_dev)
print("Variance:", variance)
# --- IGNORE ---
```

```python
#logical operations with numpy
arr = np.array([10, 15, 20, 25, 30])
greater_than_20 = arr > 20
less_than_25 = arr < 25
print("Elements greater than 20:", arr[greater_than_20])
print("Elements less than 25:", arr[less_than_25])  
# --- IGNORE ---
```
