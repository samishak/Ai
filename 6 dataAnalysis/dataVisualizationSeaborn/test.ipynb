```python
import seaborn as sns
```

```python
#basic plotting with seaborn
tips=sns.load_dataset('tips')
tips
```

```python
#create  a scatter plot
import matplotlib.pyplot as plt
sns.scatterplot(data=tips, x='total_bill', y='tip', hue='day', style='time', size='size')
plt.title('Scatter Plot of Total Bill vs Tip')
plt.xlabel('Total Bill')
plt.ylabel('Tip Amount')
plt.show()  
```

```python
#line plot
sns.lineplot(data=tips, x='size', y='total_bill')
plt.title('Line Plot of Total Bill vs Tip')
plt.xlabel('size')
plt.ylabel('total bill')
plt.show()
```

```python
#categorical plot 
#bar plot
sns.barplot(data=tips, x='day', y='total_bill', hue='smoker')
plt.title('Bar Plot of Total Bill by Day and Smoker Status')
plt.xlabel('Day')
```

```python
#create box plot
sns.boxplot(data=tips, x="day", y='total_bill')
plt.title('Box Plot of Total Bill by Day')
plt.ylabel('Total Bill')
plt.show()

```

```python
# violin plot
sns.violinplot(x='day', y='total_bill', data=tips)

plt.title('Violin Plot of Total Bill by Day')
plt.ylabel('Total Bill')
plt.show()
```

```python
#histograms 
sns.histplot(tips['total_bill'],bins=10,kde=True)
```

```python
#kde plot
sns.kdeplot(tips['total_bill'],shade=True)
```

```python
#PairPlot
sns.pairplot(tips)
```

```python
#heatmap
corr=tips[['total_bill','tip','size']].corr()
corr
```

```python
sns.heatmap(corr,annot=True,cmap="coolwarm")
```
