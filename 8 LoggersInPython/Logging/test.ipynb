```python
import logging
#We use loggers in Python to record what a program is doing while it runs,
# especially to debug, monitor, and maintain applications in a clean and professional way.
logging.basicConfig(
    level=logging.DEBUG,
    format="%(asctime)s - %(name)s - %(levelname)s - %(message)s",
    datefmt="%Y-%m-%d %H:%M:%S",
    handlers=[
        logging.FileHandler("app.log", mode="w"),
        logging.StreamHandler()  # prints to the notebook/terminal
    ],
    force=True
)
#log file msg
logging.debug("this is a logging message")
logging.info("this is an info msg")
logging.warning("Warning msg")
logging.error("error msg")
logging.critical("critical msg")
```
